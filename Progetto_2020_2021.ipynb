### 5. Consider only the most recent data point of the times dataset. Compute the number of male and female students for each country.¶

```python
time_mry = time.loc[time.groupby('university_name')['year'].idxmax()]
time_mry
```

```python
time_mry_mfc = time_mry.dropna(subset = ['female_male_ratio','num_students']) #elimino le righe che contengono valori nulli nelle colonne num_students e female_male_ratio
time_mry_mfc = time_mry_mfc[['country','university_name','num_students','female_male_ratio']]
```

```python
rate = time_mry_mfc.set_index('university_name')['female_male_ratio'].str.split(pat=':', expand = True).reset_index() #splitto i due valori separati dal simbolo ':'
rate.dropna(inplace=True)
rate.rename(columns = {0:'female_percentage', 1:'male_percentage'}, inplace = True) #rinomino le colonne

rate['female_percentage'] = pd.to_numeric(rate['female_percentage'])
rate['male_percentage'] = pd.to_numeric(rate['male_percentage'])
```

```python
rate
```

```python
time_mry_mfc.dtypes
```

```python
#time_mry_mfc = time_mry_mfc.astype(str)
#time_mry_mfc.dtypes
```

```python
time_mry_mfc['num_students'] = time_mry_mfc['num_students'].astype(str).str.replace(',','') #sostituisco le virgole con spazio vuoto
time_mry_mfc['num_students'] = pd.to_numeric(time_mry_mfc['num_students'])
```

```python
time_mry_mfc
```

```python
merge_time_rate = time_mry_mfc.merge(rate, how='right', on = 'university_name') #mergio le due tabelle con un right join rispetto alle università
```

```python
merge_time_rate.dtypes
```

```python
#calcolo le proporzioni di maschi e femmine per ciascuna data e ciascuna università
merge_time_rate['num_female'] = (merge_time_rate['female_percentage']*merge_time_rate['num_students'])/100
merge_time_rate['num_male'] = (merge_time_rate['male_percentage']*merge_time_rate['num_students'])/100
```

```python
#approssimo i numeri ad degli interi
merge_time_rate['num_female'] = merge_time_rate['num_female'].round()
merge_time_rate['num_male'] = merge_time_rate['num_male'].round()
```

```python
#raggruppo i dati in base al paese e faccio la somma
total_female = merge_time_rate.groupby('country')['num_female'].sum()
total_male = merge_time_rate.groupby('country')['num_male'].sum()
```

```python
total_female
```

```python
total_male
```
